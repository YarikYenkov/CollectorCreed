```python
picfiles = []
for f in os.listdir():
    if f.endswith('.jpg') or f.endswith('.png'):
        picfiles.append(f)
print(picfiles)
```

```python
picd = {}
for pic in picfiles:
    img = cv2.imread(pic)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    picd[pic] = img
    fig, ax = plt.subplots(1, 1)
    ax.imshow(img)
```

```python
img2 = cv2.imread('Steve.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2, (40, 80))
```

```python
img[390:390 + img2.shape[0], 900:900 + img2.shape[1], :] = img2
# img[:img2.shape[0], :img2.shape[1], :] = img2
fig, ax = plt.subplots(1, 1, figsize = (20, 20))
ax.imshow(img)
```

```python
for x in range(img2.shape[0]):
    for y in range(img2.shape[1]):
        if np.any(img2[x, y, :]):
            img[390 + x, 900 + y, :] = img2[x, y, :]
```
